```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#Normal dist
x=np.random.normal(loc=5,scale=2,size=(2000)) #--> give an array with nd
#loc: mean ,the centre of vals,scale: mean of values,size: space 1d with variables
 
#uniform dist
c=np.random.uniform(low=-3,high=2,size=20)

#cdf
def cdf_uniform(x):
    if x<0:
        return 0
    elif x<1:
        return x
    else:
        return 1




#cdf for uniform
f=[cdf_uniform(x) for x in c]



#cdf for normal
d=[]



print(c.std() , c.mean())
print(c)
```
